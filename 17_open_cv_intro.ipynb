```python
import cv2
from matplotlib import pyplot as plt

grey_img = cv2.imread('images/test.jpg', flags=0) #2 Chanell Grey images. 2-D arr
color_img = cv2.imread('images/test.jpg', flags=1) # 3 Chanell BGR images 3-D arr
p_color_image = plt.imread('images/test.jpg')


# !!! opencv reads images as BGR 
# !!! plt reads images as RGB 

# print(grey_img)
# print(color_img)
# print( p_color_image)

plt.imshow(color_img)
cv2.imshow("Grey Image", grey_img) #not responding window
cv2.imshow("Color Image", color_img) #not responding window
# to solve : 
cv2.waitKey(10000) #time to destroy all windows. but 0  
cv2.destroyAllWindows()

```

plt assumes images as rgb but cv2 assumes images as bgr

so for example if you want to use plt.imshow method you'll need to convert it from bgr to rgb

plt:
[[[
    r,g,b
]]] 

opencv : 
[[[
    b,g,r
]]]

```python
plt.imshow(cv2.cvtColor(color_img,cv2.COLOR_BGR2RGB))
```

GLOB

```python
import glob

path = 'images/*'

for file in glob.glob(path):

    origin_img = cv2.imread(file)
    print(file, origin_img.shape)
    cv2.imshow("Origin img", origin_img)
    cv2.waitKey(3000)
    cv2.destroyAllWindows()

    cvt_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
    print(file, cvt_img.shape)
    cv2.imshow("Color img", cvt_img)
    cv2.waitKey(3000)
    cv2.destroyAllWindows()
    

```
